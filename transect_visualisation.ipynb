```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

experiment = 'Exp1'

all_data = pd.read_csv('total_data.csv')
all_data = all_data[all_data['Experiment']==experiment]
for x in np.sort(all_data['Distance'].unique()):
    transect_x = x
    transect_data = all_data[all_data['Distance']==x]
    for height in transect_data['Height'].unique():
        transect_z = transect_data[transect_data['Height']==height]['Height']
        transect_y = transect_data[transect_data['Height']==height]['Peak_Dist']
        transect_conc = transect_data[transect_data['Height']==height]['Concentration']

        plt.scatter(transect_z,transect_y,s=transect_conc/100,c='b')
    plt.title('Downstream Distance = '+str(x)+'m')
    plt.xlabel('Height')
    plt.ylabel('Distance from Peak Concentration')
    plt.show()
```
